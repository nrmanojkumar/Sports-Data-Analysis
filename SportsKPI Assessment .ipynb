```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'SportsKPI_Kabaddi_Dataset_for_Assessment (1) (1).xlsx'  
data = pd.read_excel(file_path)
```

a)	Distribution of Raids (counts and percentage) for any Raider (use Raider Name column)

```python
raider_counts = data['Raider_Name'].value_counts()
raider_percentage = (raider_counts / raider_counts.sum()) * 100
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Bar plot for counts
sns.barplot(x=raider_counts.index, y=raider_counts.values, ax=ax1, palette="viridis")
ax1.set_title("Raid Counts for Each Raider")
ax1.set_xlabel("Raider Name")
ax1.set_ylabel("Number of Raids")
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)

# Pie chart for percentages
ax2.pie(raider_percentage, labels=raider_percentage.index, autopct='%1.1f%%', colors=sns.color_palette("viridis", len(raider_percentage)))
ax2.set_title("Raid Distribution Percentage by Raider")

plt.tight_layout()
plt.show()
```

b)	Distribution of Tackles (counts and percentage) for any Defender (use Defender_1_Name column)

```python
# Calculate tackle counts and percentage for each defender
defender_counts = data['Defender_1_Name'].value_counts()
defender_percentage = (defender_counts / defender_counts.sum()) * 100
```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Bar plot for counts
sns.barplot(x=defender_counts.index, y=defender_counts.values, ax=ax1, palette="magma")
ax1.set_title("Tackle Counts for Each Defender")
ax1.set_xlabel("Defender Name")
ax1.set_ylabel("Number of Tackles")
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=90)

# Pie chart for percentages
ax2.pie(defender_percentage, labels=defender_percentage.index, autopct='%1.1f%%', colors=sns.color_palette("magma", len(defender_percentage)))
ax2.set_title("Tackle Distribution Percentage by Defender")

plt.tight_layout()
plt.show()
```

c)	Plot Attacking Skills Distribution for any Raider.

```python
selected_raider = 'MONU' 
raider_data = data[data['Raider_Name'] == selected_raider]
```

```python
# Calculate counts of attacking skills for the selected raider
attacking_skills_counts = raider_data['Attacking_Skill'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=attacking_skills_counts.index, y=attacking_skills_counts.values, palette="plasma")
plt.title(f"Attacking Skills Distribution for {selected_raider}")
plt.xlabel("Attacking Skill")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

d)	Plot Defensive Skills Distribution for any Defender.

```python
# Select a specific defender 
selected_defender = 'MANOJ'  
defender_data = data[data['Defender_1_Name'] == selected_defender]
```

```python
# Calculate counts of defensive skills for the selected defender
defensive_skills_counts = defender_data['Defensive_Skill'].value_counts()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=defensive_skills_counts.index, y=defensive_skills_counts.values, palette="coolwarm")
plt.title(f"Defensive Skills Distribution for {selected_defender}")
plt.xlabel("Defensive Skill")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
